```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('mail_data.csv')
```

```python
df
```

```python
data = df.where((pd.notnull(df)),)
```

```python
data.head(20)
```

```python
data.info()
```

```python
data.shape
```

```python
data.loc[data['label']== 'spam','label',] = 0
data.loc[data['label']== 'ham','label',] = 1
```

```python
data.info()
```

```python
X= data['text']
Y = data['label']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state = 3) 
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
pred = model.predict(X_train_features)
accu = accuracy_score(Y_train,pred)
```

```python
print ("Accuracy :",accu)
```

```python
test_data= model.predict(X_test_features)
accuracy=accuracy_score(Y_test,test_data)
print("Accuracy :",accuracy)
```
